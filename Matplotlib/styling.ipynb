```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,11)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')
ax.plot(x,x**2,label='X vs X^2')

ax.legend(loc=(1.1,0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color = '#b4bd3c',label = 'X vs X',lw=5,ls='--')

ax.plot(x,x+1,color = 'purple',label = 'X vs X + 1',lw=2,ls='-.')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#matplotlib markers
ax.plot(x,x+1,color = 'purple',label = 'X vs X + 1',lw=2,ls='--',marker='s',ms=10,markerfacecolor='red',markeredgewidth=4,markeredgecolor='orange')
ax.plot(x,x,color = 'red',label = 'X vs X',lw=2,ls='--',marker='+',ms=20)

ax.legend()
```
