```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(a,b)
ax.set_xlim(0,8)
ax.set_ylim(0,8000)
ax.set_xlabel('A')
ax.set_ylabel('B')
ax.set_title('Возведение в степень 4')

ax1 = fig.add_axes([1,1,2,2])
ax1.plot(a,b)
ax1.set_xlim(1,2)
ax1.set_ylim(0,50)
ax1.set_xlabel('A')
ax1.set_ylabel('B')
ax1.set_title('Увеличенный фрагмент');
```
