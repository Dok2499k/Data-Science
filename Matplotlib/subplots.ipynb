```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0,0].plot(x,y)
axes[0,1].plot(a,b)
axes[1,0].plot(x,y)
axes[1,1].plot(a,b)

fig.suptitle('Figure Level Title')
plt.tight_layout()

# fig.subplots_adjust(wspace=0.9,hspace=0.5)
```
