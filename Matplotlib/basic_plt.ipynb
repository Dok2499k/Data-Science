```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
plt.plot(x,y)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.xlim(0,6) #для своего рода увеличения графика
plt.ylim(0,15)
plt.title('Название графика');
# plt.show()
```
