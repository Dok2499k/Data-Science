```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
sns.rugplot(data=df,x='salary')
```

```python
# sns.set(style='darkgrid')
sns.displot(data=df,x='salary',bins=20,color='red',kde=True)#,edgecolor='blue',linewidth=4)
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=['age'])
```

```python
sns.displot(data=sample_ages,x='age',kde=True,bins=30)
```

```python
sns.kdeplot(data=sample_ages,x='age',clip=[0,100],fill=True,bw_adjust=0.6)
```
