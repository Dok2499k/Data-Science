```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='division')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='level of education')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='division',hue='level of education',palette='Set2')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=df,x='level of education',y='salary',estimator='mean',errorbar='sd')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=df,x='level of education',y='salary',estimator='mean',errorbar='sd',
            hue='division')
plt.legend(loc=(1.05,0.5))
```

```python
df1 = pd.read_csv('StudentsPerformance.csv')
```

```python
df1.head()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df1,y='math score',x='test preparation course')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df1,y='reading score',x='test preparation course')
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.boxplot(data=df1,y='reading score',x='parental level of education',
            hue='test preparation course',palette='Set2')
plt.legend(loc=(1.1,0.5))
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df1,x='reading score',y='parental level of education',
               hue='test preparation course',palette='Set2')
plt.legend(loc=(1.1,0.5))
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df1,x='reading score',y='parental level of education',
               hue='test preparation course',palette='Set2',split=True,inner='quartile')
plt.legend(loc=(1.1,0.5))
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df1,x='reading score',y='parental level of education',
               split=True,inner='quartile')
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df1,x='reading score',y='parental level of education',
               split=True,inner='stick')
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df1,x='reading score',y='parental level of education',bw=0.1) # чем меньше bw тем больше шума ловишь на данных
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(data=df1,x='math score',y='gender',size=2,hue='test preparation course',dodge=True)
plt.legend(loc=(1.1,0.5))
```
